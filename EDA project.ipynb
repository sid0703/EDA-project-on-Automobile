```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("Automobile_data.csv")
df
```

```python
df=pd.read_csv("Automobile_data.csv")
df
```

```python
df.describe()
```

```python
df[['engine-size','price']].hist(figsize=(7,5),bins=6)
plt.show()
```

```python
m=df["engine-size"].mean()
me=df["engine-size"].median()
sns.distplot(df["engine-size"])
plt.axvline(m,color='r')
plt.axvline(me,color='y')
plt.grid(True)
```

```python
sns.countplot(data=df,x="body-style",hue="fuel-type")
plt.grid(True)
```

```python
plt.figure(1)
plt.subplot(2,2,1)
df['engine-type'].value_counts(normalize=True).plot(figsize=(10,8),kind='bar',color='purple')
plt.title("Number of engine type frequency diagram")
plt.xlabel("Number of engine type")
plt.ylabel("engine type")

plt.subplot(2,2,2)
df['drive-wheels'].value_counts(normalize=True).plot(figsize=(10,8),kind='bar',color='yellow')
plt.title("Number of drive wheels frequency diagram")
plt.xlabel("Number of drive wheels")
plt.ylabel("drive wheels")

plt.subplot(2,2,3)
df['fuel-type'].value_counts(normalize=True).plot(figsize=(10,8),kind='bar',color='blue')
plt.title("Number of fuel type frequency diagram")
plt.xlabel("Number of fuel type")
plt.ylabel("fuel type")


plt.subplot(2,2,4)
df['body-style'].value_counts(normalize=True).plot(figsize=(10,8),kind='bar',color='green')
plt.title("Number of body style frequency diagram")
plt.xlabel("Number of body style type")
plt.ylabel("body style type")
plt.tight_layout()
plt.show()
```

```python
explode=[0,0.1,0]
df['drive-wheels'].value_counts(normalize=True).plot(figsize=(10,8),kind='pie',autopct='%.2f%%',explode=explode)
plt.show()
```

```python
df.corr()
```

```python
cor=df.corr()
plt.figure(figsize=(7,5))
sns.heatmap(cor,annot=True,cmap='cividis')
```

```python
plt.figure(figsize=(23,10))
sns.boxplot(data=df,x='body-style',y='price')
plt.show()
```

```python
sns.pairplot(df[["city-mpg","engine-size","price","highway-mpg","fuel-type"]],hue="fuel-type",diag_kind="hist")
plt.show()
```
